# Aquisição de dados 
Código para a aquisição de dados de forma consolidada a partir da base de dados fornecida. 

Uso de:

*   Biblioteca rpy2 - para converter o arquivo .rds para o python
*   Biblioteca os - para manipulação de arquivos em diretório
*   Biblioteca pandas - para construir data frame consolidado e exportá-lo em CSV.

Aquisição de todos os processos da pasta 'falencias'

```
from os import listdir
from os.path import isfile, join
import rpy2.robjects as ro
import pandas as pd

readRDS = ro.r['readRDS']

PATH = '/content/drive/My Drive/6-Processamento de Linguagem Natural em decisões judiciais/dataset1/falencias'
onlyfiles = [f for f in listdir(PATH) if isfile(join(PATH, f))] #Lista de arquivos na pasta

NumeroDoProcesso = []
html = []
ProcessoEletronico = []

for x in onlyfiles:
  df_r = readRDS(PATH + '/' + x)
  df = ro.conversion.rpy2py(df_r)
  NumeroDoProcesso.append(df[0])
  html.append(df[1])
  ProcessoEletronico.append(df[2])

df_final = pd.DataFrame({"NumeroDoProcesso" : NumeroDoProcesso, 
                         "html" : html,
                         "ProcessoEletronico" : ProcessoEletronico})

df_final.head()
```

Exportar em CSV

```
df_final.to_csv('alencias.csv', index = False)
```

Aquisição de todos os processos da pasta 'recuperacoes_judiciais' e exportação em CSV

```
from os import listdir
from os.path import isfile, join
import rpy2.robjects as ro
import pandas as pd

readRDS = ro.r['readRDS']

PATH = '/content/drive/My Drive/6-Processamento de Linguagem Natural em decisões judiciais/dataset1/recuperacoes_judiciais'
onlyfiles = [f for f in listdir(PATH) if isfile(join(PATH, f))] #Lista de arquivos na pasta

NumeroDoProcesso = []
html = []
ProcessoEletronico = []

for x in onlyfiles:
  df_r = readRDS(PATH + '/' + x)
  df = ro.conversion.rpy2py(df_r)
  NumeroDoProcesso.append(df[0])
  html.append(df[1])
  ProcessoEletronico.append(df[2])

df_final = pd.DataFrame({"NumeroDoProcesso" : NumeroDoProcesso, 
                         "html" : html,
                         "ProcessoEletronico" : ProcessoEletronico})
```

```
df_final.to_csv('recuperacoes_judiciais.csv', index = False)
```
